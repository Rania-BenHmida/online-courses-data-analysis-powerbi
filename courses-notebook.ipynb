```python
import pandas as pd
import numpy as np

# Load dataset
path = "/kaggle/input/online-courses/Online_Courses.csv"
df = pd.read_csv(path)

print("Initial shape:", df.shape)
df.head()
```

```python
def get_used_columns(site):
    subset = df[df["Site"] == site]
    used = []

    for col in df.columns:
        # If at least one non-null value exists for this column in this site
        if subset[col].notna().sum() > 0:
            used.append(col)
    
    return used

print("Future Learn:", get_used_columns("Future Learn"))
print("Coursera:", get_used_columns("Coursera"))
print("Udacity:", get_used_columns("Udacity"))
print("Simplilearn:", get_used_columns("Simplilearn"))
```

```python
future = df[df["Site"] == "Future Learn"].copy()
coursera = df[df["Site"] == "Coursera"].copy()
udacity = df[df["Site"] == "Udacity"].copy()
simplilearn = df[df["Site"] == "Simplilearn"].copy()
```

# **1. Cleaning the courses from 'Future Learn'**

```python
future.shape
```

```python
# Vérification des doublons
print(f"Duplicated rows: {future.duplicated().sum()}")
```

```python
# Vérification des valeurs manquantes
print(future.isnull().sum())
```

```python
# Find columns that are completely empty for Future Learn
cols_to_remove = [col for col in future.columns if future[col].isna().sum() == len(future)]

# Drop them
future_clean = future.drop(columns=cols_to_remove)

print("Removed columns:")
print(cols_to_remove)

print("Shape before:", future.shape)
print("Shape after:", future_clean.shape)
```

```python
future_clean.head(5)
```

```python
future_clean = future_clean.drop(columns=[
    "Course Title",
    "Course URL",
    "Course Short Intro",
    "FAQs",
    "What's include"
])
```

```python
print(future_clean.columns.tolist())
```

```python
future_clean = future_clean.rename(columns={
    "Topics related to CRM": "Skills",
    "ExpertTracks": "Course Type",
    "Premium course" : "Price",
    "Weekly study": "Weekly hours"
})
```

```python
print(future_clean.columns.tolist())
```

```python

# Remove leading numbers and spaces from Course Type
future_clean["Course Type"] = future_clean["Course Type"].str.replace(r'^\d+\s*', '', regex=True)

future_clean["Course Type"] = future_clean["Course Type"].replace("Degrees", "Degree")

# Check the result
future_clean["Course Type"].unique()

```

```python
# Create Category column from Skills
future_clean["Category"] = future_clean["Skills"].apply(
    lambda x: x.split('/')[0].strip() if isinstance(x, str) and '/' in x else x
)
```

```python
import re

def futurelearn_duration_to_hours(row):
    duration_str = str(row["Duration"]).lower()
    weekly_hours = row.get("Weekly hours", np.nan)
    
    # Default weekly hours if missing
    try:
        weekly_hours = float(weekly_hours)
    except:
        weekly_hours = 10  # default
    
    # Extract number of weeks
    w = re.findall(r'(\d+)\s*week', duration_str)
    if w:
        total_hours = int(w[0]) * weekly_hours
        return total_hours
    return np.nan

# Apply to future_clean
future_clean["Duration"] = future_clean.apply(futurelearn_duration_to_hours, axis=1)

# Drop Weekly hours column
future_clean = future_clean.drop(columns=["Weekly hours"])

# Check result
future_clean["Duration"].head(10)
```

```python
def clean_price(x):
    if isinstance(x, str):
        x = x.lower().replace("free", "0")        # convert "Free" to 0
        x = ''.join(c for c in x if c.isdigit() or c=='.')  # keep digits and decimal
    try:
        return float(x)
    except:
        return np.nan

future_clean["Price"] = future_clean["Price"].apply(clean_price)
```

# **2. Cleaning the courses from 'Coursera'**

```python
coursera.shape
```

```python
# Vérification des doublons
print(f"Duplicated rows: {coursera.duplicated().sum()}")
```

```python
# Find columns that are completely empty for Future Learn
cols_to_remove2 = [col for col in coursera.columns if coursera[col].isna().sum() == len(coursera)]

# Drop them
coursera_clean = coursera.drop(columns=cols_to_remove2)

print("Removed columns:")
print(cols_to_remove2)

print("Shape before:", coursera.shape)
print("Shape after:", coursera_clean.shape)
```

```python
print(coursera_clean.columns.tolist())
```

```python
import re

def coursera_duration_to_hours(x):
    x = str(x).lower()
    # hours
    h = re.findall(r'(\d+(\.\d+)?)\s*hour', x)
    if h:
        return float(h[0][0])
    # minutes
    m = re.findall(r'(\d+(\.\d+)?)\s*min', x)
    if m:
        return float(m[0][0])/60
    # weeks
    w = re.findall(r'(\d+(\.\d+)?)(\s*-\s*(\d+))?\s*week', x)
    if w:
        if w[0][3]:  # range
            avg = (float(w[0][0]) + float(w[0][3])) / 2
        else:
            avg = float(w[0][0])
        return avg * 10  # assume 10 hours/week if not specified
    # months
    mo = re.findall(r'(\d+(\.\d+)?)\s*month', x)
    if mo:
        return float(mo[0][0]) * 4 * 10  # 4 weeks/month, 10 hours/week as default
    return np.nan
```

```python
coursera_clean["Duration"] = coursera_clean["Duration"].apply(coursera_duration_to_hours)
```

# **3. Cleaning the courses from 'Udacity'**

```python
udacity.shape
```

```python
# Vérification des doublons
print(f"Duplicated rows: {udacity.duplicated().sum()}")
```

```python
# Find columns that are completely empty for Future Learn
cols_to_remove3 = [col for col in udacity.columns if udacity[col].isna().sum() == len(udacity)]

# Drop them
udacity_clean = udacity.drop(columns=cols_to_remove3)

print("Removed columns:")
print(cols_to_remove3)

print("Shape before:", udacity.shape)
print("Shape after:", udacity_clean.shape)
```

```python
print(udacity_clean.columns.tolist())
```

```python
udacity_clean = udacity_clean.rename(columns={
    "Program Type": "Course Type",
    "School": "Category",
    "Number of Reviews": "Number of viewers"
})

# Check
udacity_clean.columns
```

```python
import numpy as np

price_cols = ["6-Month access", "4-Month access", "3-Month access", "5-Month access", "2-Month access"]

def clean_price_udacity(x):
    if isinstance(x, str):
        x = ''.join(c for c in x if c.isdigit() or c=='.')  # keep only digits and dot
    try:
        return float(x)
    except:
        return np.nan  # treat empty or invalid as NaN

# Apply cleaning
for col in price_cols:
    udacity_clean[col] = udacity_clean[col].apply(clean_price_udacity)

# Create total Price (sum, ignoring NaNs)
udacity_clean["Price"] = udacity_clean[price_cols].sum(axis=1, skipna=True)

# Optional: drop original access columns
udacity_clean = udacity_clean.drop(columns=price_cols)

# Check
udacity_clean[["Price"]].tail(10)
```

```python
udacity_clean.columns
```

```python
cols_to_drop = ["Prequisites", "What you learn", "Related Programs", "Monthly access"]
udacity_clean = udacity_clean.drop(columns=cols_to_drop)

# Check remaining columns
udacity_clean.columns
```

```python
def udacity_duration_to_hours(x):
    x = str(x).lower()
    # weeks range
    w = re.findall(r'(\d+)\s*-\s*(\d+)\s*week', x)
    if w:
        avg = (int(w[0][0]) + int(w[0][1])) / 2
        return avg * 10
    # weeks
    w2 = re.findall(r'(\d+)\s*week', x)
    if w2:
        return int(w2[0]) * 10
    # months
    mo = re.findall(r'(\d+)\s*month', x)
    if mo:
        return int(mo[0]) * 4 * 10
    return np.nan
```

```python
udacity_clean["Duration"] = udacity_clean["Duration"].apply(udacity_duration_to_hours)
```

# **5. Rejoining files and adjusting columns**

```python
# List columns for each site
site_dfs = {
    "Future Learn": future_clean,
    "Coursera": coursera_clean,
    "Udacity": udacity_clean,
    "Simplilearn": simplilearn_clean
}

for site, df_site in site_dfs.items():
    print(f"\n=== Columns in {site} ===")
    for col in df_site.columns:
        print(col)
```

```python
# Define final columns
final_cols = ["Unnamed: 0", "Title", "URL", "Short Intro", "Category", "Sub-Category",
              "Course Type", "Skills", "Level", "Courses", "Rating", "Number of viewers",
              "Duration", "Price", "Language", "Subtitle Languages", "Site", "Instructors"]
```

```python
# Ensure all columns exist in each DataFrame
for df in [future_clean, coursera_clean, udacity_clean, simplilearn_clean]:
    for col in final_cols:
        if col not in df.columns:
            df[col] = np.nan
```

```python
# Reorder columns
future_clean = future_clean[final_cols]
coursera_clean = coursera_clean[final_cols]
udacity_clean = udacity_clean[final_cols]
simplilearn_clean = simplilearn_clean[final_cols]
```

```python
# Combine
all_courses_clean = pd.concat([future_clean, coursera_clean, udacity_clean, simplilearn_clean],
                              ignore_index=True)
```

```python
all_courses_clean.info()
```

```python
# Columns to clean
numeric_cols = ["Courses", "Rating", "Number of viewers"]

# Function to keep only numbers and convert to float
def to_numeric(x):
    if isinstance(x, str):
        # Remove everything except digits and dot
        x = ''.join(c for c in x if c.isdigit() or c == '.')
    try:
        return float(x)
    except:
        return np.nan

# Apply to each column
for col in numeric_cols:
    all_courses_clean[col] = all_courses_clean[col].apply(to_numeric)
```

```python
# Optional: check results
print(all_courses_clean[numeric_cols].dtypes)
print(all_courses_clean[numeric_cols].head(10))
```

```python
# Save cleaned dataset to CSV
all_courses_clean.to_csv("all_courses_clean.csv", index=False)

print("Saved all_courses_clean.csv successfully!")
```
